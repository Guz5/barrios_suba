```python
import geopandas as gpd
import matplotlib.pyplot as plt
import json
```

```python
suba = gpd.read_file("suba.geojson")

suba
```

```python
suba.to_crs(epsg=9377, inplace=True)

suba
```

```python
suba.plot()
```

```python
mapa_2 = suba.plot(
    column = "nombre",
    figsize = (15,15),
    edgecolor = "black",
    legend = True,
    cmap = "tab20",
    legend_kwds = {"loc": "lower left", "ncol": 2}
)

legend = mapa_2.get_legend()
legend.set_title("Leyenda")
legend.set_bbox_to_anchor((1.2, 1))
```
